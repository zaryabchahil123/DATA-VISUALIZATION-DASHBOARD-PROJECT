```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ['Sequence_Name', 'mcg', 'gvh', 'alm', 'mit', 'erl', 'pox', 'vac', 'nuc', 'Class']
df = pd.read_csv('../data/yeast.data', sep=r'\s+', names=column_names)
df.head()
```

```python
class_counts = df['Class'].value_counts()
total = class_counts.sum()
threshold = 0.03 * total
large_classes = class_counts[class_counts >= threshold]
small_classes_sum = class_counts[class_counts < threshold].sum()
if small_classes_sum > 0:
    large_classes['Other'] = small_classes_sum
plt.figure(figsize=(8, 8))
plt.pie(large_classes, labels=large_classes.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("muted"), pctdistance=0.75, wedgeprops=dict(width=0.35, edgecolor='w'))
plt.title("Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['mcg'], bins=30, kde=True, color='#3498db')
plt.title("Distribution of mcg")
plt.xlabel("mcg Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values(by='mcg').reset_index(drop=True)
sns.lineplot(data=df_sorted, x=df_sorted.index, y='gvh', color='#e74c3c')
plt.title("Trend of gvh across sorted mcg")
plt.xlabel("Sequence Index (sorted by mcg)")
plt.ylabel("gvh Value")
plt.show()
```

```python
class_avg = df.groupby('Class')['alm'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=class_avg, x='Class', y='alm', hue='Class', palette="viridis", legend=False)
plt.title("Average alm per Class")
plt.xlabel("Class")
plt.ylabel("Average alm")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mcg', y='gvh', hue='Class', palette="Set2")
plt.title("Relationship between mcg and gvh")
plt.xlabel("mcg")
plt.ylabel("gvh")
plt.legend(title="Class", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Class', y='nuc', hue='Class', palette="husl", legend=False)
plt.title("Data Spread of nuc across Classes")
plt.xlabel("Class")
plt.ylabel("nuc")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Features")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values(by='mcg').reset_index(drop=True)
df_sorted['mit'].plot(kind='area', alpha=0.5, color='mediumseagreen')
plt.title("Cumulative Trend of mit over sorted mcg")
plt.xlabel("Sequence Index (sorted by mcg)")
plt.ylabel("mit Value")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='erl', hue='erl', palette="Set1", legend=False)
plt.title("Frequency of erl Categories")
plt.xlabel("erl")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Class', y='vac', hue='Class', palette="muted", legend=False)
plt.title("Distribution and Probability Density of vac by Class")
plt.xlabel("Class")
plt.ylabel("vac")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sizes = df['vac'] * 500 + 50
sns.scatterplot(data=df, x='mcg', y='gvh', hue='Class', size=sizes, sizes=(20, 500), palette="Set1", alpha=0.7)
plt.title("mcg vs gvh Bubble Chart (Size = vac)")
plt.xlabel("mcg")
plt.ylabel("gvh")
plt.legend(title="Class", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
